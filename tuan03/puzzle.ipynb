Câu 3: Cho bàn cờ độ rộng 3 x 3 chứa 8 con số từ 1 đến 8 và 1 ô trống xuất hiện ngẫu nhiên. Hãy
viết thuật toán xếp thứ tự các con số tăng dần từ 1 đến 8 theo chiều từ trái sang phải và từ trên xuống dưới.

Biểu diễn bàn cờ qua thuật toán:


```python
start = (4, 8, 1,
         6, 3, 0,
         2, 7, 5)

goal = (1, 2, 3,
        4, 5, 6,
        7, 8, 0)
```

Phép toán để thực hiện:

```python
def get_next(state):
    neighbors = []

    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                x, y = i, j

    moves = [(-1,0),(1,0),(0,-1),(0,1)] 

    for dx, dy in moves:
        nx, ny = x + dx, y + dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = [list(row) for row in state]
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            neighbors.append(tuple(tuple(row) for row in new_state))

    return neighbors
```

Chạy thuật toán theo BFS:

```python
from collections import deque

def BFS(start, goal):
    q = deque([start])
    parent = {to_tuple(start): None}

    while q:
        current = q.popleft()

        if current == goal:
            break

        for nxt in get_next(current):
            t = to_tuple(nxt)
            if t not in parent:
                parent[t] = current
                q.append(nxt)

    # truy vết đường đi
    path = []
    cur = goal
    while cur is not None:
        path.append(cur)
        cur = parent[to_tuple(cur)]
    return path[::-1]
```

Chạy thuật toán theo DFS:

```python
def DFS(start, goal):
    stack = [start]
    parent = {start: None}

    while stack:
        current = stack.pop()
        if current == goal:
            break
        for nxt in get_next(current):
            if nxt not in parent:
                parent[nxt] = current
                stack.append(nxt)

    path = []
    node = goal
    while node:
        path.append(node)
        node = parent[node]

    return path[::-1]
```
