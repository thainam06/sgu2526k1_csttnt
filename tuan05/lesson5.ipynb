```python
n,k=map(int, input().split())
a=[]
for _ in range(n):
    a.append(input().strip())
print(f'{n} {k}')
for i in a:
    print(i)
```

```python
x=[0]*k
result=[]

def Try(i,start):
    for j in range(start,n):
        x[i]=j
        if i==k-1:
            combo=[]
            for t in range(k):
                combo.append(a[x[t]])
            result.append(combo)
        else:
            Try(i+1,j+1)
```

```python
Try(0,0)

print(len(result))
for i in result:
    print(i)
```

### Greedy


```python
n=int(input("nhập kích thước bàn cờ"))
print(n)
board=[[0]*n for _ in range(n)]
col=[False]*n
diag1=[False]*(2*n)
diag2=[False]*(2*n)
for i in board:
    print(i)
```

```python
def Greedy():
    for r in range(n):
        placed=False
        for c in range(n):
            if not col[c] and not diag1[r-c+n] and not diag2[r+c]:
                board[r][c]=1
                col[c]= True
                diag1[r-c+n]=True
                diag2[r+c]=True
                
                placed=True
                break
        if not placed:
            return 0
    return 1
```

```python
status=Greedy()
if status == 0:
    print("0")
else:
    print("1")
    for i in board:
        print(*i)
```

## Bài 3

```python
n,x0,y0=map(int,input().split())
board=[[0]*n for _ in range(n)]
dx=[-2,-2,-1,-1,1,1,2,2]
dy=[-1,1,-2,2,-2,2,-1,1]
found=False
def inside(x,y):
    return 0<=x<n and 0<=y<n
def Try(step,x,y):
    global found
    if step==n*n:
        for i in range(8):
            nx=x+dx[i]
            ny=y+dy[i]
            if nx==x0 and ny==y0:
                found=True
                return
        return
    for i in range(8):
        nx=x+dx[i]
        ny=y+dy[i]
        if inside(nx,ny) and board[nx][ny]==0:
            board[nx][ny]=step+1
            Try(step+1,nx,ny)
            if found:
                return
            board[nx][ny]=0
board[x0][y0] = 1
Try(1,x0,y0)
if not found:
    print("0")
else:
    print("1")
    for i in board:
        print(*i)
```

### Greedy

```python
data = """
9
1 2
1 4
1 5
2 3
3 4
3 7
4 5
4 6
5 6
5 9
6 8
6 9
""".strip().split('\n')

n = int(data[0])

adj = [[] for _ in range(n+1)]

for line in data[1:]:
    u,v = map(int,line.split())
    adj[u].append(v)
    adj[v].append(u)
color=[0]*(n+1)
for u in range(1,n+1):
    used=set()
    for v in adj[u]:
        if color[v]!=0:
            used.add(color[v])
    c=1
    while c in used:
        c+=1
    color[u]=c
k=max(color)
print(k)
groups=[[] for _ in range(k+1)]
for i in range(1,n+1):
    groups[color[i]].append(i)
for i in range(1,k+1):
    print(*groups[i])
```

### greedy + sắp thứ tự

```python
edges = [
("California","Nevada"),
("California","Arizona"),

("Nevada","Utah"),
("Nevada","Arizona"),

("Utah","Arizona"),
("Utah","Colorado"),

("Arizona","Colorado"),

("Colorado","Nebraska"),
("Colorado","Oklahoma"),

("Nebraska","Kansas"),

("Kansas","Oklahoma"),

("Oklahoma","Texas"),
("Oklahoma","Arkansas"),

("Texas","Arkansas"),
("Texas","Louisiana"),

("Arkansas","Louisiana"),
("Arkansas","Mississippi"),

("Mississippi","Louisiana"),
("Mississippi","Georgia"),

("Georgia","Florida"),
("Georgia","Mississippi")
]
```

```python
adj = {}

for u,v in edges:

    if u not in adj:
        adj[u] = []

    if v not in adj:
        adj[v] = []

    adj[u].append(v)
    adj[v].append(u)
```

```python
order=sorted(adj.keys(),key=lambda x:len(adj[x]),reverse=True)
color={}
for city in order:
    used=set()
    for nex in adj[city]:
        if nex in color:
            used.add(color[nex])
    c=1
    while c in used:
        c+=1
    color[city]=c
```

```python
k = max(color.values())
print(k)

groups = [[] for _ in range(k+1)]

for city in color:
    groups[color[city]].append(city)

for i in range(1,k+1):
    print(*groups[i])
```

### backtracking(tối ưu hoá)

```python
nodes = list(adj.keys())
def valid(u,c):
    for v in adj[u]:
        if color.get(v)==c:
            return False
    return True
def Try(i,k):
    if i==n:
        return True
    u=nodes[i]
    for c in range(1,k+1):
        if valid(u,c):
            color[u]=c
            if Try(i+1,k):
                return True
            del color[u]
    return False
for k in range(1,n+1):
    color={}
    if Try(0,k):
        print(k)
        groups=[[] for _ in range(k+1)]
        for city in color:
            groups[color[city]].append(city)
        for i in range(1,k+1):
            print(*groups[i])
        break
```
